```python
from bs4 import BeautifulSoup
import requests as rq
import pandas as pd

year = int(input("시작할 년도(숫자만): "))
month = int(input("시작할 달(숫자만): "))
team = input("검색할 팀(현대캐피탈=1/IBK기업은행=2): ")
teamName = {'1':'현대캐피탈', '2':'IBK기업은행'}
teamCode = {'1':1005, '2':2006}

startYear = year
startMonth = month

gameDate = []
gameTime = []
gameAway = []
gameHome = []
gameStadium = []
gameRound = []
gameBoard = []

for i in range(2):
    while month <= 12:
        print(f'{year}년 {month}월 수집 시작')
        url = f'https://sports.news.naver.com/volleyball/schedule/index?month={month}&year={year}&teamCode={teamCode[team]}'
        print(url)
        res = rq.get(url)
        soup = BeautifulSoup(res.text,'html.parser')
        
        YearMonth = soup.find_all('em', class_='now_selected')
        if int(YearMonth[0].text[-4:]) != year or int(YearMonth[1].text[-2:]) != month:
            break
            
        noGame = soup.find_all('div', class_='sch_tb nogame')
        noGame2 = soup.find_all('div', class_='sch_tb2 nogame2')

        try:
            soup.find('div', class_='sch_tb nogame selected').decompose()
        except:
            pass
        try:
            soup.find('div', class_='sch_tb2 nogame2 selected').decompose()
        except:
            pass

        for i in range(len(noGame)):
            soup.find('div', class_='sch_tb nogame').decompose()
        for j in range(len(noGame2)):
            soup.find('div', class_='sch_tb2 nogame2').decompose()

        gameDay = soup.find_all('div', class_='sch_tb')
        gameDay2 = soup.find_all('div', class_='sch_tb2')
        for day in gameDay2:
            gameDay.append(day)
        if gameDay == []:
            break

        for k in gameDay:
            gameDate.append(str(year)+'.'+k.find('span', class_='td_date').text[:-4])
            gameTime.append(k.find('span', class_='td_hour').text)
            gameAway.append(k.find('span', class_='team_lft').text)
            gameHome.append(k.find('span', class_='team_rgt').text)
            gameStadium.append(k.find('span', class_='td_stadium').text)
            gameRound.append(k.find('span', class_='td_round').text)
            gameBoard.append(k.find('span', class_='td_broad').text.strip('\n                                '))

        month += 1
    if gameDay == []:
        break
    year += 1
    month = 1
year -= 1
    
subject = []
startDate = []
startTime = []
endDate = []
endTime = []
allDayEvent = [] # True or False
description = []
location = [] 
private = [] # True or False

for m in range(len(gameDate)):
    subject.append(gameAway[m]+' vs '+gameHome[m])
    startDate.append(gameDate[m])
    startTime.append(gameTime[m])
    allDayEvent.append('False')
    description.append(gameRound[m]+' '+gameAway[m]+' vs '+gameHome[m]+' (중계방송사: '+gameBoard[m]+')')
    location.append(gameStadium[m])
    private.append('False')
endDate = startDate
endTime = startTime

gameDict = {'subject':subject, 'start date':startDate, 'start time':startTime, 'end date':endDate,
           'end time':endTime, 'all day event':allDayEvent, 'description':description,
           'location':location, 'private':private}


gameDf = pd.DataFrame(gameDict)
display(gameDf)

gameDf.to_csv(teamName[team]+str(startYear)+'_'+str(startMonth)+'.csv', encoding="utf-8-sig", index=False)
   
```
